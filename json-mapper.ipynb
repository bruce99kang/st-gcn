```python
for frame_number in data.keys():
    try:
        frame_number = int(frame_number)
    except Exception as e:
        continue
    id_name = list(data.get(str(frame_number)).keys())[2]
    no_detect1 = no_detect.copy()
    for n,idx in enumerate(no_detect):
        for id_detect in data.get(str(frame_number)).get(id_name):
            if idx.get('id_part') == id_detect.get('id_part'):
                #print(idx.get('id_part'))
                no_detect1[n] = id_detect
                break
    data.get(str(frame_number)).update({id_name:no_detect1})
```
